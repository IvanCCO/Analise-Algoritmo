```python
# Imports
import matplotlib.pyplot as plt
import time
import numpy as np
import matplotlib as mpl
import random
from memory_profiler import memory_usage
import sys
import timeit
from functools import partial
sys.setrecursionlimit(999_999_999)
```

```python
def bubble_sort(arr):
    n = len(arr)
    # Percorre todos os elementos do array
    for i in range(n):
        # Últimos i elementos já estão no lugar certo, então não precisamos mais verificar eles
        for j in range(0, n - i - 1):
            # Percorre o array de 0 a n-i-1
            # Troca se o elemento atual for maior que o próximo
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr
```

```python
def create_arr(num : int) -> list[int]:
    arr = []
    for i in range(1,num):
        j = i * 100
        new_array = list(range(j))
        random.shuffle(new_array)
        arr.append(new_array)
    return arr
```

```python
if __name__ == "__main__":
    plt.ion()  # Ativando o modo interativo
    x = list(range(1, 100))  # Valores para o eixo x (tamanho do array)
    y = []

    plt.show(block=False)

    for i in x:
        arr = create_arr(i)

        inicio = time.time()
        bubble_sort(arr)
        fim = time.time()

        y.append(round(fim - inicio, 2))

        # Plotar o gráfico em tempo real
        plt.clf()
        plt.plot(x[:len(y)], y, marker='o')
        plt.title('Desempenho do Bubble Sort')
        plt.xlabel('Tamanho do Array')
        plt.ylabel('Tempo (segundos)')
        plt.draw()
        plt.pause(0.1)  # Pausa para permitir a atualização do gráfico

    plt.ioff()  # Desativar o modo interativo após a conclusão do loop
    plt.show()
```
